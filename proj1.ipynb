```python
c = pd.read_csv("C:/Users/lpfac/Documents/git/CMSC491FinalProj/2010-2013/Education_and_Youth__2010-2013_.csv")
c
```

```python
results = pd.DataFrame(columns=('ColNames','Correlation'))
results
```

```python
crimIter = iter(f)
next(crimIter)
for i in crimIter:
    econIter = iter(b)
    next(econIter)
    eduIter = iter(c)
    next(eduIter)
    for j in econIter:
        corVal= np.corrcoef(x= f[i], y=b[j]) 
        temp = pd.DataFrame({"ColNames":[str(i+j)], "Correlation":[corVal[0][1]]}) 
        results= results.append(temp,ignore_index = True)
    for k in eduIter:
        corVal= np.corrcoef(x= f[i], y=c[k]) 
        temp = pd.DataFrame({"ColNames":[str(i+k)], "Correlation":[corVal[0][1]]}) 
        results= results.append(temp,ignore_index = True)
```

```python
sortedVals = results.sort_values(by='Correlation')
```
